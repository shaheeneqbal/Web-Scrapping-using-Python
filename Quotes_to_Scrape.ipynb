```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
allQuotes = []

for i in range(1,6):
    response = requests.get(f"http://quotes.toscrape.com/page/{i}/")
    soup = BeautifulSoup(response.content,'html.parser')
    quotes = soup.find_all('div',class_ = 'quote')
    
    for quote in quotes:
        texts = quote.find('span',class_ = 'text').text
        author = quote.find('small').text
        tags = quote.find('div',class_ = 'tags').find_all('a')
    
        tags_list = []
        for tag in tags:
            tags_list.append(tag.text)
        
        allQuotes.append([author,texts,tags_list])
```

```python
df = pd.DataFrame(allQuotes, columns = ['Author Name','Quote','Tags'])
display(df)
```
